```python
#Homework 1
# Data preprocessing

import pandas as pd
import numpy as np
df=pd.read_csv('titanic.csv')
df
```

```python
df=pd.read_csv('titanic.csv', names=['PassengerId','Survived','Pclass','Name','Sex','Age','SibSp','Parch','Ticket','Fare','Cabin','Embarked'],
               na_values='?')
df
```

```python
df.drop('Name', axis=1)
```

```python
df.drop('Name', axis=1, inplace=True)
```

```python
df.drop('Ticket', axis=1)
```

```python
df.drop('Cabin', axis=1)
```

```python
df.drop('Cabin', axis=1, inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder

number = LabelEncoder()

df['Sex'] = number.fit_transform(df['Sex'].astype(str))
```

```python
# Converting binary Nominal Variable Gender to numeric
df['Sex'].replace({'Male':1, 'Female':2}, inplace=True)
print(df)
```

```python
#print misssing value rate (object)
for i in df.select_dtypes(include=np.object).columns.tolist():
    n_miss = df[[i]].isnull().sum()
    perc = n_miss / df.shape[0] * 100
    print('> %s, Missing: %d (%.1f%%)' % (i, n_miss, perc))
```

```python
df.fillna(df.mean())
```

```python
df.fillna(df.mean(), inplace=True)
```

```python
#Normalization and standardization 
#StandardScalerscaler

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
d=scaler.fit_transform(df.Sex.values.reshape(-1, 1))

df['standard_s']=d
df
```

```python
#min-max sacle
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range=(-3,3))
d=scaler.fit_transform(df.Sex.values.reshape(-1, 1))

df['min_max_s']=d
df
```

```python
#max-abs sacle
from sklearn.preprocessing import MaxAbsScaler

scaler = MaxAbsScaler()
d=scaler.fit_transform(df.Sex.values.reshape(-1, 1))

df['max_abs_s']=d
df
```

```python
#robost sacle
from sklearn.preprocessing import RobustScaler

robust = RobustScaler(quantile_range = (0.1,0.9))
d=robust.fit_transform(df.Sex.values.reshape(-1, 1))

df['robost_s']=d
df
```

```python
#max norm
df['B_norm_max']=df.Sex/df.Sex.max()
df
```

```python
#norm_l1
df['Sex_norm_l1']=df.Sex/df.Sex.sum()
df
```

```python
#norm l2

df['Sex_norm_l2']=df.Sex/sum(df.Sex**2)
df
```

```python
# Q2.Python Fibonacci series Program using For Loop
# Fibonacci series will start at 0 and travel upto below number

Number = int(input("\nPlease Enter the Range Number: "))

# Initializing First and Second Values of a Series
First_Value = 0
Second_Value = 1
           
# Find & Displaying Fibonacci series
for Num in range(0, Number):
           if(Num <= 1):
                      Next = Num
           else:
                      Next = First_Value + Second_Value
                      First_Value = Second_Value
                      Second_Value = Next
           print(Next)
```

```python
# Q3. Queue

import queue

p_queue = queue.PriorityQueue()
p_queue.put((2, "A"))
p_queue.put((1, "B"))
p_queue.put((3, "C"))
print(p_queue.get())
```

```python
#sorting queue
import queue
q1 = queue.Queue()

q1.put((5, "A"))
q1.put((2, "B"))
q1.put((1, "C"))
q1.put((6, "D"))
q1.put((4, "E"))
q1.put((3, "F"))

while (q1.empty() == False): 
    print(q1.queue[0], end = " ")  
    q1.get()
```
